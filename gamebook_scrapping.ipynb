```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Website URL (using books.toscrape.com as a demo)
url = "https://books.toscrape.com/"

# Send request to website
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# Lists to store extracted data
game_names = []
prices = []
ratings = []

# Find all products (treating them as games)
games = soup.find_all("article", class_="product_pod")

for game in games:
    # Game name
    name = game.h3.a["title"]
    game_names.append(name)
    
    # Game price
    price = game.find("p", class_="price_color").text
    prices.append(price)
    
    # Rating (One, Two, Three, Four, Five)
    rating = game.p["class"][1]
    ratings.append(rating)
    # Create a DataFrame
df = pd.DataFrame({
    "Game Name": game_names,
    "Price": prices,
    "Rating": ratings
})

# Show proper table
df.to_csv("gramebook.csv",index=False)
```
